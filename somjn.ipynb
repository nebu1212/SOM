```r
# Installieren der benötigten Bibliotheken
install.packages("kohonen")
```

```r
# Laden der benötigten Bibliotheken
library(kohonen)
```

```r
# Zeigen des Iris-Datensatzes (bekannt aus der Vorlesung)
str(iris)
```

```r
# Entfernen der Species-Werte aus dem Datensatz sowie Umwandlung in eine Matrix
iris_as_matrix <- as.matrix(scale(iris[,-5]))
```

```r
# Anzeigen der Matrix
print(iris_as_matrix)
```

```r
# Setzte X- und Y-Dimensionen der Ausgangsmatrix
# Sowohl x_max als auch y_max müssen > 0 sein
x_max = 5
y_max = 5
```

```r
# Setze die Form oder Topologie der Ausgangsmatrix
# Wähle dabei aus "hexagonal", "rectangular"
topo_setting <- "hexagonal"
```

```r
# Setze die Funktion, die im Training für die Anpassung der Nachbarn gewählt werden soll
# Wähle dabei #zwischen "bubble" und "gaussian"
neighbourhood_setting = "bubble"
```

```r
# Setze den Toroidal-Parameter (ob die Oberseite-Unterseite und Rechts-Links miteinander verbunden sind)
# Wähle dabei zwischen TRUE und FALSE
toroidal_setting = FALSE
```

```r
# Erstelle SOM-Grid
som_grid <- somgrid(xdim = x_max, ydim = y_max, topo = topo_setting, neighbourhood.fct = neighbourhood_setting, toroidal = toroidal_setting )
```

```r
# Anzeigen des Grids
# Beinhaltet die Koordinaten der einzelnen Felder der Ausgangsmatrix
# Anzahl der Einträge ist x_max * y_max
print(som_grid)
```

```r
# Setze Seed, damit ein Datensatz wieder reproduziert werden kann
set.seed(2021)
```

```r
# Setze die Trainingswiederholungen
iterations = 100
```

```r
# Rufe die som-Funktion auf, um das Modell zu trainieren
SOM_model <- som(X = iris_as_matrix, grid = som_grid, rlen = iterations)
```

```r
# 1. Plot: Anzahl der Treffer je Feld der Ausgangsmatrix
# Graue Felder = keine Zuteilung eines Datensatzes
# Je heller, desto mehr Datensätze sind einem Feld zugeteilt
plot(SOM_model, type= "counts")
```

```r
# 2. Plot: Heatmaps der Verteilung der einzelnen Eigenschaften
# Knoten mit Datensätzen, in denen die Eigenschaft hoch ist, werden hell und mit größeren Zahlen angegeben
plot(SOM_model, type="property", property = getCodes(SOM_model)[,1], main = colnames(iris)[1])
plot(SOM_model, type="property", property = getCodes(SOM_model)[,2], main = colnames(iris)[2])
plot(SOM_model, type="property", property = getCodes(SOM_model)[,3], main = colnames(iris)[3])
plot(SOM_model, type="property", property = getCodes(SOM_model)[,4], main = colnames(iris)[4])
```
